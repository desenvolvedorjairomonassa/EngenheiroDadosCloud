```python
import pandas as pd
import numpy as np
import random

source = '../input/enem-2019/DADOS/MICRODADOS_ENEM_2019.csv'

total_linhas = sum(1 for l in open(source,encoding="ISO-8859-1") )#conta quantas linhas tem o data set
print(f'Total de Linhas: {total_linhas}')

dados = pd.read_csv(source,encoding="ISO-8859-1",delimiter=";",
                    usecols=['NU_NOTA_MT','NU_NOTA_CH','NU_NOTA_CN','TP_SEXO','SG_UF_RESIDENCIA','NU_IDADE','CO_ESCOLA',
                             'IN_GESTANTE','IN_DEFICIENCIA_FISICA','IN_CEGUEIRA','NO_MUNICIPIO_PROVA',
                             'NO_MUNICIPIO_NASCIMENTO','NO_MUNICIPIO_RESIDENCIA'])
dados.shape
```

```python
dados.head(10)
```

```python
#Sobre o ENEM 2019, qual é a média da nota em matemática apenas para as alunas do sexo Feminino?
dados.loc[dados['TP_SEXO'] == 'F']['NU_NOTA_MT'].mean()
```

```python
#Qual é a média da nota em Ciências Humanas para os alunos do sexo masculino que residem no estado de São Paulo?
dados.loc[(dados['TP_SEXO'] == 'M')& (dados['SG_UF_RESIDENCIA']=='SP')]['NU_NOTA_CH'].mean()
```

```python
#Qual é a cidade (município de residência) que possui a MAIOR nota em Ciências Naturais?
maiores = dados.groupby('NO_MUNICIPIO_RESIDENCIA')['NU_NOTA_CN'].max().sort_values(ascending=False).head(1)
maiores
#maiores entras 4 cidades 'Montes Claros','Governador Valadares','Ouro Preto','Ipatinga'
dados.loc[dados['NO_MUNICIPIO_RESIDENCIA'].isin(['Montes Claros','Governador Valadares','Ouro Preto','Ipatinga'])].groupby('NO_MUNICIPIO_RESIDENCIA')['NU_NOTA_CN'].max().sort_values(ascending=False)
#maiores.loc[maiores['NO_MUNICIPIO_RESIDENCIA'] == 'Montes Claros']
```

```python
# MAIOR nota em Ciências Naturais?
dados['NU_NOTA_CN'].max()
```

```python
#Qual é a média de idade dos alunos que NASCERAM na cidade de Natal?
dados.loc[dados['NO_MUNICIPIO_NASCIMENTO']=='Natal']['NU_IDADE'].mean()
```

```python
#Qual é o código da escola que obteve a maior nota MÉDIA em matemática?
dados.groupby('CO_ESCOLA')['NU_NOTA_MT'].mean().sort_values(ascending=False).head(3)
```

```python
#Quantas pessoas que moram no estado do Rio de Janeiro do sexo masculino obtiveram nota em matemática maior do que 600?
dados.loc[(dados['TP_SEXO']=='M')& (dados['SG_UF_RESIDENCIA']=='RJ') & (dados['NU_NOTA_MT']> 600) ].shape[0]
```

```python
#Quantas pessoas nascidas em Recife fizeram a prova do ENEM nessa mesma cidade?
dados.loc[(dados['NO_MUNICIPIO_NASCIMENTO']=='Recife')& (dados['NO_MUNICIPIO_PROVA']=='Recife')       ].shape[0]
```

```python
#Quantas mulheres gestantes que moram no estado de SP fizeram a prova de Ciências Humanas do ENEM em 2019?
dados.loc[(dados['SG_UF_RESIDENCIA']=='SP')& (dados['IN_GESTANTE'] == 1)&(dados['NU_NOTA_CH']>=0)].shape
```

```python
#Qual é a nota média em matemática dos alunos cegos do sexo feminino residentes em Belo Horizonte?
dados.loc[(dados['NO_MUNICIPIO_RESIDENCIA']=='Belo Horizonte')&(dados['TP_SEXO']=='F')&(dados['IN_CEGUEIRA'] == 1)]['NU_NOTA_MT'].mean()
```

```python
#Qual é a nota média em Ciências Humanas dos alunos que moram no estado de Santa Catarina e possuem deficiência física?
dados.loc[(dados['SG_UF_RESIDENCIA']=='SC')&(dados['IN_DEFICIENCIA_FISICA']==1)]['NU_NOTA_CH'].mean()
```
